```python
#finding top wines prices

#note that you set all the prices to 80 above
top_wine = wine_data_df.sort_values(["price"], ascending=False)
#              ^            ^           ^            ^                
#          dataframe      function    column      ascending        
#            to             name        to           or            
#           sort                     sort by      descending      

top_wine.head()
```

```python
#how to create country level dataframes 
usDF = wine_data_df.loc[wine_data_df['country'] == 'US']
#          ^         ^              ^            
#      dataframe   locate         where to look*
#                 something

#* ill go over this more during our session
usDF.head()
```

```python
#how to create a sub dataframe 
usTrimmed = usDF[["country", "points", "price", "variety"]]
usTrimmed.head()
```

```python
#how to use the sort by and unique function 
usPriceSort = usTrimmed.sort_values("points", ascending = False)
usPriceSort.head(15)
```

```python
#unique gives us all of the.... unique values in a certain column
usPriceSort['variety'].unique()
#in this case a long list of all the varieties
```

```python
#slightly different to unique is value_counts
usPriceSort['variety'].value_counts()
#it returns all the varieties... but also the amount of each... these come in the form of key value pairs so you can do...
```

```python
usPriceSort['variety'].value_counts()['Syrah']
#and this will give us the count of how many Syrah's there are
```

```python
#how to separate values from one column into two columns 
split = usPriceSort["variety"].str.split(" ", n = 1, expand = True)
usSplit = usPriceSort
usSplit['variety1'] = split[0]
usSplit['variety2'] = split[1]
usSplit.head()
```
