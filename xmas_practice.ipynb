```python
import pandas as pd
import numpy as np
```

```python
#file to load 
file_to_load = "xmas_practice.csv"
```

```python
#read the wine data

wine_data_df = pd.read_csv(file_to_load)
wine_data_df
```

```python
#calculate total wine entries

wine_count = wine_data_df["country"].count()

wine_count
```

```python
#dictionary of countries and number of wines per each 
wine_data_df['country'].value_counts().to_dict()
```

```python
#are there any empty cells?
wine_data_df.isnull()
```

```python
len(wine_data_df["country"].unique())
```

```python
average_wine_price = wine_data_df["price"].mean()
average_wine_price
```

```python
expensive_wine = wine_data_df["price"] = 80
expensive_wine
```

```python
#finding top wines prices
top_wine = wine_data_df.sort_values(["price"], ascending=False)

top_wine.head()
```
